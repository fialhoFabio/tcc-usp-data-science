```python
import pandas as pd
import matplotlib.pyplot as plt
import locale
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, ccf, grangercausalitytests, acf
from statsmodels.tsa.api import VAR


locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')
df = pd.read_csv('../data/df_analise.csv').set_index('Data')
df.index = pd.to_datetime(df.index, format='%b/%Y')
df = df.asfreq(pd.infer_freq(df.index))
```

```python
for col in df.columns:
    if col != 'Dolar' and col != 'IBOVESPA':
        # Estacionaridade
        result_adfuller = adfuller(df['IPCA'], autolag='AIC')
        print(col + ' Stationary: ' + str(result_adfuller[1] < 0.05))

        # Autocorrelacao
        print(acf(x=df[col], nlags=24))
        plot_acf(x=df[col], lags=24, title='ACF: ' + col)
        plt.title('ACF: ' + col, fontsize=10)  # Reduced font size
        plt.show()

        # Cross correlation
        result_ccf = ccf(df['Dolar'], df[col], adjusted=True)[:60]
        print(result_ccf[:60])
        plt.plot(result_ccf, label=col)
        plt.title('CCF: ' + col + ' x Dolar', fontsize=10)  # Reduced font size
        plt.axhline(0, color='red', linestyle='--', linewidth=1)
        plt.show()

        # VAR SELECT ORDER
        var_model = VAR(df[['Dolar', col]])
        lag_order = var_model.select_order(maxlags=12)
        print(lag_order.summary())

        # VAR FIT - LAG SEGUINDO AIC
        result_var = 0
        if col=='IPCA':
            result_var = var_model.fit(maxlags=1)
        elif col=='INPC':
            result_var = var_model.fit(maxlags=1)
        elif col=='IGPM':
            result_var = var_model.fit(maxlags=8)
        elif col=='IGPDI':
            result_var = var_model.fit(maxlags=8)
        elif col=='IGP10':
            result_var = var_model.fit(maxlags=8)

        print(result_var.summary())

        # VAR Causality
        cause = result_var.test_causality(col, 'Dolar', kind='f')
        print(cause.summary())

        # IRF
        irf = result_var.irf(12)
        print(irf)
        irf.plot(impulse='Dolar', response=col, figsize=(6, 4))  # Adjusted figure size
        plt.title(f"IRF: Dolar -> {col}", fontsize=10)  # Reduced font size
        plt.show()
        dolar_index = 0
        x_index = 1

        # Respostas de x a choques no Dlar
        response_x_to_dolar = irf.orth_irfs [:, dolar_index, x_index]

        for month, value in enumerate(response_x_to_dolar):
            print(f"Ms {month}: {value:.5f}")

        # FEVD
        fevd = result_var.fevd(24)

        # Plot the FEVD
        print(fevd)
        fevd.plot(figsize=(6, 4))  # Adjusted figure size
        plt.title("Dolar -> " + col, fontsize=10)  # Reduced font size
        plt.xlabel("Steps")
        plt.ylabel("Proportion")
        plt.legend(loc='best')

        for level in [0.25, 0.5, 0.75]:
            plt.axhline(level, color='gray', linestyle='--', linewidth=0.8)

        plt.show()
```
